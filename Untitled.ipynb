# Data Fetcher

```python
import os
import wget
from datetime import datetime, timedelta, date

print('Cek keberadaan direktori ...')
directory = 'fetched'
if os.path.isdir(directory):
    print('Direktori ditemukan !')
else:
    print('Membuat direktori ...')
    os.mkdir(directory)

print('Fetching dimulai ...')
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/'
url += 'csse_covid_19_data/csse_covid_19_daily_reports/'
today = date.today().strftime('%m-%d-%Y')
x = '01-22-2020'
while x < today:
    if os.path.isfile(directory+'/'+x+'.csv'):
        pass
    else:
        wget.download(url+x+'.csv', directory+'/'+x+'.csv')
        print(' -> Sukses unduh '+x+'.csv')
    x = (datetime.strptime(x, '%m-%d-%Y') + timedelta(days=1)).strftime('%m-%d-%Y')
print('Fetching selesai !')

print('Periksa keberadaan file dengan ekstensi .tmp ...')
files_in_directory = os.listdir(directory)
all_tmp = [file for file in files_in_directory if file.endswith(".tmp")]
if len(all_tmp) == 0:
    print('Tidak ada file dengan extensi .tmp !')
else:
    print(f'Ditemukan {len(all_tmp)} buah file dengan extensi .tmp')
    for file in all_tmp:
        path_to_file = os.path.join(directory, file)
        os.remove(path_to_file)
    print('Pembersihan direktori selesai !')
print('Tahap sinkronisasi data selesai !')
```
