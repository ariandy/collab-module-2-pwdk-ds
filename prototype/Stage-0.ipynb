Data di atas masih dalam keadaan mentah. Ada beberapa hal yang harus kita lakukan pada hasil merger dari banyak dataset tersebut.

**Step-1**, ada beberapa kolom yang redundan. Yang mana kolom redundan ini adalah sebagai berikut.
- 'Province/State' <---> 'Province_State'
- 'Country/Region' <---> 'Country_Region'
- 'Lat' <---> 'Latitude'
- 'Long_' <---> 'Longitude'
- 'Last Update' <---> 'Last_Update'

Tugas awal kita adalah,
- Lakukan sinkronisasi data dari kedua kolom tersebut. (1a)
- Hapus salah satu kolom yang redundant, **dengan catatan**, pastikan kedua kolom yang redundant haruslah tersinkronisasi valuenya. (1b)

**Step-2**
- Isi NaN pada kolom 'FIPS' dengan 0.0 dan isi NaN pada 'Admin2' dengan 'Unknown'. Kolom ini saya sebut sebagai "US-only related columns", karena 2 kolom ini hanya berlaku dan diperuntukkan untuk regional US. Oleh karena itu, kolom 'FIPS' yang bernilai 0.0 dan juga kolom 'Admin2' yang bernilai 'Unknown' menujukkan bahwa baris tersebut bukan dari regional US (2a)
- Drop column 'Combined_Key'. Pada dokumentasi teknis yang terbaru, kolom ini sudah tidak tertera lagi. (Meskipun nantinya kolom ini akan di-regenerate kembali untuk keperluan tertentu). (2b)

**Step-3**
- Isi NaN pada column 'Province_State' dengan 'Unknown'. (3a)
- Ubah 'Mainland China' menjadi 'China' (agar tidak redundant semisal diubah menjadi time series). (3b)
- Rapikan posisi columns. (3c)

**Step-4**,membuat semua format penanggalan pada 'Last_Update' menjadi MM/DD/YYYY HH:mm:ss (format 24 jam).
Konvensi pengisian tanggal pada dataset ini sudah beberapa kali melakukan pergantian format penanggalan. Oleh karenanya, format penanggalan yang ada pada kolom 'Last_Update' menjadi heterogen. Untuk itu, kita perlu ubah formatnya mengikuti konvensi terakhir yang digunakan oleh dokumentasi teknis.

Step-4 ini akan dipecah menjadi beberapa Act, yaitu bagian-bagian yang (pada awalnya) dijalankan secara gradual (per cell), sampai semua kondisinya terpenuhi, sehingga selanjutnya bisa dijalankan secara utuh.

- **Act-0** : Buat 1 kolom baru (lu_temp) dengan nilai default 'N'.
- **Act-1** : Snip terlebih dahulu semua format yang di dapat pertama kali. Ini dapat dilakukan dengan cara mengambil beberapa baris pertama yang nilai 'lu_temp'nya adalah 'N'
- **Act-2** : Tulis secara gradual semua ekspresi Regex yang ditemukan pada Act-1
- **Act-3** : Lakukan filter dengan Regex matcher. Jika belum menemukan nilai True dan counter bernilai 0, kembali ke Act-1. Jika sudah menemukan nilai True dan counter bernilai 0, lanjut ke Act-Last.
- **Act-Last** : Aktifkan '%%script false --no-raise-error' pada Act-1 dan Act-3

Di tahap ini, Act sudah bisa dijalankan seperti biasanya. Lalu selanjutnya yang harus dilakukan.
- **Converter** : Membuat converter untuk setiap Regex pattern yang ditemukan.
- **Last Filter** : Salin script dari Act-3, namun dengan '%%script false --no-raise-error' yang dinonaktifkan. Lakukan filter terakhir dengan script ini untuk memastikan semua telah terkonversi.
- **Date Splitter** : Buat satu kolom (Date) dimana valuenya didapat dari date yang berada dari kolom 'Last_Update'. Ini berguna untuk pengolahan Data pada bagian selanjutnya.

**Step-N**, kita harus menghandle columns seperti 'Confirmed', 'Deaths', 'Recovered', 'Lat', 'Long_', 'Active', 'Incidence_Rate', 'Case-Fatality_Ratio', sampai akhirnya siap untuk diolah.

# Step-1

```python
# 1a dan 1b sudah disatukan dalam fungsi column_eliminator()

origin = ['Province/State', 'Country/Region', 'Last Update', 'Latitude', 'Longitude']
target = ['Province_State', 'Country_Region', 'Last_Update', 'Lat', 'Long_']

if len(origin) == len(target):
    for i in range(len(origin)):
        column_eliminator(df, origin[i], target[i])
```

# Step-2

```python
# 2a
df['FIPS'].fillna(value=0.0, inplace=True)
df['Admin2'].fillna(value='Unknown', inplace=True)

# 2b
df.drop('Combined_Key', axis=1, inplace=True)
```

# Step-3

Sebelum melakukan mengisi NaN pada kolom ini, ada beberapa hal yang perlu diperhatikan.

Ada 3 baris data dari 'Province_State' yang bernilai 'None'.<br>
Data ini tidak saya ubah, dikarenakan saya tidak tahu apakah 'None' ini adalah sebuah nama kota (seperti kota municipality di Italy atau seperti nama sebuah daerah di Azerbaijan), atau 'None' yang dimaksud ini adalah kata 'None' dalam bahasa Inggris.

```python
df[df['Province_State']=='None']
```

Dan ada beberapa baris yang bernilai 'Unknown'.

```python
df[df['Province_State']=='Unknown']
```

```python
# 3a
df['Province_State'].fillna('Unknown', inplace=True)

# 3b
df['Country_Region'] = df['Country_Region'].replace('Mainland China', 'China')

# 3c
df = df[['Admin2', 'Province_State', 'Country_Region', 'Last_Update', 'Lat', 'Long_',
   'Confirmed','Recovered', 'Deaths', 'Active', 'Incidence_Rate', 'Case-Fatality_Ratio']]
```

### Act-2

```python
# 1/28/20 23:00 (Pattern A)
a = '^([1-9]|1[0-2])/([1-9]|[12][0-9]|3[01])/20 (00|[0-9]|1[0-9]|2[0-3]):([0-9]|[0-5][0-9])$'
# 2020-07-17 22:34:48 (Pattern B) ---> Ini adalah pattern yang kita inginkan
b = '^202[0-3]-(0[1-9]|1[0-2])-(0[1-9]|[12][0-9]|3[01]) (0[0-9]|1[0-9]|2[0-3]):([0-5][0-9]):([0-5][0-9])$'
# 2020-03-06T14:23:04 (Pattern C)
c = '^202[0-3]-(0[1-9]|1[0-2])-(0[1-9]|[12][0-9]|3[01])T(0[0-9]|1[0-9]|2[0-3]):([0-5][0-9]):([0-5][0-9])$'
# 1/22/2020 17:00 (Pattern D)
d = '^[1-9]/([1-9]|[12][0-9]|3[01])/20[0-9]{2} (00|[0-9]|1[0-9]|2[0-3]):([0-9]|[0-5][0-9])$'
pattern = [a, b, c, d]
```

### Act-3

```python
%%script false --no-raise-error

# Buat 1 kolom baru dengan nilai default 'N'.
df['lu_temp'] = 'N'

for i in range(len(pattern)):
    pattern_checker_report(df, pattern[i])

print(len(df[df['lu_temp']=='N'])) # memastikan bahwa sudah tidak ada lagi value 'N' pada kolom 'lu_temp'
```

### Act-Last

Aktifkan '%%script false --no-raise-error' pada Act-1 dan Act-3

### Converter

```python
df.drop('lu_temp', axis=1, inplace=True) # drop kolom 'lu_temp'
higher_order_converter(df, pattern[0], convert_a)
higher_order_converter(df, pattern[2], convert_c)
higher_order_converter(df, pattern[3], convert_a)
```

### Last Filter

```python
# %%script false --no-raise-error

# Buat 1 kolom baru dengan nilai default 'N'.
df['lu_temp'] = 'N'

for i in range(len(pattern)):
    pattern_checker_report(df, pattern[i])

print(len(df[df['lu_temp']=='N'])) # memastikan bahwa sudah tidak ada lagi value 'N' pada kolom 'lu_temp'
```

### Date Splitter

```python
splitter = df.Last_Update.str.split(expand=True)
df['Date'] = splitter[0]
df.drop('lu_temp', axis=1, inplace=True) # drop kolom 'lu_temp'
df['Last_Update'] = pd.to_datetime(df['Last_Update'])
```
