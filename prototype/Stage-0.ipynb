# Merging the CSV

```python
import glob
import pandas as pd

all_filenames = [i for i in glob.glob('fetched/*.csv')]
merged_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
merged_csv.to_csv("merged.csv", index=False, encoding='utf-8-sig')
```

Data di atas masih dalam keadaan mentah. Ada beberapa hal yang harus kita lakukan pada hasil merger dari banyak dataset tersebut.

**Step-1**, ada beberapa kolom yang redundan. Yang mana kolom redundan ini adalah sebagai berikut.
- 'Province/State' <---> 'Province_State'
- 'Country/Region' <---> 'Country_Region'
- 'Lat' <---> 'Latitude'
- 'Long_' <---> 'Longitude'
- 'Last Update' <---> 'Last_Update'

Tugas awal kita adalah melakukan sinkronisasi data dari kedua kolom tersebut.

**Step-2**, hapus salah satu kolom yang redundant, **dengan catatan**, pastikan kedua kolom yang redundant haruslah tersinkronisasi valuenya.

**Step-3**, isi NaN pada kolom 'FIPS' dengan 0.0 dan isi NaN pada 'Admin2' dengan '-'. Kolom ini saya sebut sebagai "US-only related columns", karena 2 kolom ini hanya berlaku dan diperuntukkan untuk regional US. Oleh karena itu, kolom 'FIPS' yang bernilai 0.0 dan juga kolom 'Admin2' yang bernilai '-' menujukkan bahwa baris tersebut bukan dari regional US

**Step-4**, drop column 'Combined_Key'. Pada dokumentasi teknis yang terbaru, kolom ini sudah tidak tertera lagi.

**Step-5**, isi NaN pada column 'Province_State' dengan 'Unknown'.

**Step-6**, rapikan posisi columns.

**Step-7**,membuat semua format penanggalan pada 'Last_Update' menjadi MM/DD/YYYY HH:mm:ss (format 24 jam).
Konvensi pengisian tanggal pada dataset ini sudah beberapa kali melakukan pergantian format penanggalan. Oleh karenanya, format penanggalan yang ada pada kolom 'Last_Update' menjadi heterogen. Untuk itu, kita perlu ubah formatnya mengikuti konvensi terakhir yang digunakan oleh dokumentasi teknis.

**Step-N**, kita harus menghandle columns seperti 'Confirmed', 'Deaths', 'Recovered', 'Lat', 'Long_', 'Active', 'Incidence_Rate', 'Case-Fatality_Ratio'

# Step-1 & Step-2

```python
column_eliminator(df, 'Province/State', 'Province_State')
column_eliminator(df, 'Country/Region', 'Country_Region')
column_eliminator(df, 'Last Update', 'Last_Update')
column_eliminator(df, 'Latitude', 'Lat')
column_eliminator(df, 'Longitude', 'Long_')
```

# Step-3 & Step-4

```python
df['FIPS'].fillna(value=0.0, inplace=True)
df['Admin2'].fillna(value='-', inplace=True)
df.drop('Combined_Key', axis=1, inplace=True)
```

# Step-5
